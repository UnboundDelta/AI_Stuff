# CustomerChurn Project

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
churn_df = pd.read_csv("Churn_Modelling.csv")

churn_df.head()
```

```python
# removing unnecessarily columns

churn_df = churn_df.drop(['RowNumber', 'CustomerId', 'Surname'], axis = 1)
```

```python
X = churn_df.drop(['Exited'], axis = 1)

Y = churn_df['Exited']
```

```python
X.head()
```

```python
Y.head()
```

```python
numerical = X.drop(['Geography', 'Gender'], axis = 1)

numerical.head()
```

```python
categorical = X.filter(['Geography', 'Gender'])

categorical.head()
```

```python
cat_numerical = pd.get_dummies(categorical, drop_first=True)
cat_numerical.head()
```

```python
# Concatenating both cat_numerical and numerical

X = pd.concat([numerical, cat_numerical], axis = 1)

X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

log_clf = LogisticRegression()
classifier = log_clf.fit(X_train, Y_train)
Y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(confusion_matrix(Y_test, Y_pred))
print(classification_report(Y_test, Y_pred))
print(accuracy_score(Y_test, Y_pred))
```

```python
# The output shows that for 81 percent of the records in the test set, logistic regression correctly
# predicted whether or not a customer will leave the bank.
```
