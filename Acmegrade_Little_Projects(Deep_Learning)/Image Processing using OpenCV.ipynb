```python
# Convert the color channels of an image from BGR to RGB

panda_img_temp = cv2.imread(r"pandas_image.jpg")
```

```python
panda_img = cv2.cvtColor(panda_img_temp, cv2.COLOR_BGR2RGB)
plt.imshow(panda_img)
```

```python
# Saving an image in different format

panda_img = cv2.imread(r"pandas_image.jpg")
cv2.imwrite(r"panda_opencv.png", panda_img)
```

```python
# Converting an image to grayscale

panda_img_temp = cv2.imread(r"pandas_image.jpg")
panda_img = cv2.cvtColor(panda_img_temp, cv2.COLOR_BGR2GRAY)
plt.imshow(panda_img)
```

```python
# Cropping an image

panda_image = cv2.imread(r"pandas_image.jpg")
y = 50
x = 50
h = 50
w = 50

crop_img = panda_image[y:y+h, x:x+w]
plt.imshow(crop_img)
```

```python
# Resizing an image
panda_img_temp = cv2.imread(r"pandas_image.jpg")
resized_img = cv2.resize(panda_img_temp, (100, 50))
plt.imshow(resized_img)
```

```python
# Applying filters

# Blur filter - It smooths an image by replacing pixels with the average value of surrounding pixels

panda_img_temp = cv2.imread(r"pandas_image.jpg")
panda_img = cv2.blur(panda_img_temp, (5, 5))
plt.imshow(panda_img)
```

```python
# Edge Detection filter

panda_img_temp = cv2.imread(r"pandas_image.jpg")
panda_img = cv2.Canny(panda_img_temp, 80, 200)
plt.imshow(panda_img)
```

```python
# Changing image orientation

from PIL import Image
panda_img = Image.open(r"pandas_image.jpg")
panda_img_ud = panda_img.transpose(Image.ROTATE_180)
display(panda_img_ud)
```

```python
panda_img_ud = panda_img.transpose(Image.FLIP_LEFT_RIGHT)
display(panda_img_ud)
```
