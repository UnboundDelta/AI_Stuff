```python
# In neural networks we only had fully connected layer, otherwise known as dense layer. With Convolutional
# neural networks, we have more operations such as the convolutional operation, max pooling, flattening and also
# a fully connected or dense layer
```

```python
# Import the libraries

from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from keras.models import Sequential
from keras.datasets import mnist
from keras.utils.np_utils import to_categorical
```

```python
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
X_train[0]
```

```python
X_train = X_train.reshape(60000, 28, 28, 1)
X_test = X_test.reshape(10000, 28, 28, 1)
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
```

```python
X_train[0]
```

```python
X_train/=255.0
X_test/=255.0
```

```python
Y_train = to_categorical(Y_train, 10)
Y_test = to_categorical(Y_test, 10)
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
# CNN Model Development

cnn = Sequential()
cnn.add(Conv2D(322, kernel_size=(3, 3), input_shape = (28, 28, 1), padding = 'same', activation='relu'))
cnn.add(MaxPooling2D())
cnn.add(Conv2D(32, kernel_size= (3, 3), padding='same', activation='relu'))
cnn.add(MaxPooling2D())
cnn.add(Flatten())
cnn.add(Dense(64, activation='relu'))
cnn.add(Dense(10, activation='softmax'))
cnn.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
print(cnn.summary())
```

```python
history_cnn = cnn.fit(X_train, Y_train, epochs = 12, verbose = 1, validation_data = (X_train, Y_train))
```

```python
plt.plot(history_cnn.history['val_accuracy'])
```
