# Web Scrapping

```python
# Importing the required libraries
from bs4 import BeautifulSoup as bs
import pandas as pd
pd.set_option('display.max_colwidth', 500)
import time
import requests
import random
```

```python
# This is the website we are using to fetch some data...Requests module helps us to ask the request to access the data..200
# shows that our request was successfully executed
page = requests.get("https://quotes.toscrape.com/")
page
```

```python
# This is the html part of that webpage
soup = bs(page.content)
soup
```

```python
# What we have to do is..first open that page and then inspect the data you want to fetch
# In the html inspect part..you will see the line which will be highlighted automatically..
# so pick up the class_='text' and use it like the following
soup.find_all(class_='text')
```

```python
# Here we don't want the starting part of every quote..so we are taking just the text part
quotes = [i.text for i in soup.find_all(class_='text')]
quotes
```

```python
# Same with authors
authors = [i.text for i in soup.find_all(class_='author')]
```

Accessing Multiple Pages

```python
# We can also access multiple pages of same website
urls = [f"https://quotes.toscrape.com/page/{i}/" for i in range(1, 11)]
urls
```

Avoiding Web Scraping Detection

```python
# Actually no website allows so many requests at one time in order to avoid a crash or so...hence we will give some pause 
# with time library to let the request happen
rate = [i/10 for i in range(10)]
time.sleep(random.choice(rate))
```

# Bringing it all Together

```python
# List of Authors and Quotes
authors = []

quotes = []

# List of URLs
urls = [f"http://quotes.toscrape.com/page/{i}/" for i in range(1,11)]

# List for Randomizing our request rate
rate = [i/10 for i in range(10)]

# Iterating through the URLS
for url in urls:
    
    # Accessing the Webpage
    page = requests.get(url)
    
    # Getting the webpage's content in pure html
    soup = bs(page.content)

    # Adding the authors and quotes to their lists
    authors.extend([i.text for i in soup.find_all(class_='author')])

    quotes.extend([i.text for i in soup.find_all(class_='text')])

    # Checking to see if we hit our required number of quotes then breaking the loop
    if len(quotes) >= 52:
        break
        
    # Randomizing our request rate    
    time.sleep(random.choice(rate))

# Creating a DataFrame to store our newly scraped information
df = pd.DataFrame()

# Storing the quotes and authors in their respective columns
df['Authors'] = authors
df['Quotes'] = quotes
```
